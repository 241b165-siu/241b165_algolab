```python
# Exercise 1
def func1(a, b):
  result = 0
  for i in range(0, b):
    result += a
  return result
```

```python
# Exercise 4
def func4(a, b):
    b.sort() # Time complexity for this function is O(Nlog N) where N = len(b)
    intersect = 0
    for x in a:
        if binary_search(b, x) >= 0:
            intersect = intersect + 1
    return intersect

def binary_search(arr, target):
    low, high = 0, len(arr) - 1
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1
    return -1
```

```python
# Sample function call: func4([1, 2, 3, 4], [2, 4, 6, 8])
a =[1, 2, 3, 4]
b =[2, 4, 6, 8]
print(func4(a,b))
import timeit
sec = timeit.timeit("func4(a,b)", globals=globals(), number=1)
print(f"Execution time: {sec*1000:.6f} ms")

#the function performs searching each element of array a in array b
```

```python
# Exercise 5
def func5(n):
    guess = 1
    while guess * guess <= n:
        if guess * guess == n:
            return guess
        guess = guess + 1
    return None
```

```python
# Sample function call: func5(16)
n = 16
print(func5(n))

import timeit
sec = timeit.timeit("func5(n)", globals=globals(), number=1)
print(f"Execution time: {sec*1000:.6f} ms")

#the function performs printing the last digit of any number
```

```python
# Exercise 6
def func8(A):
    func8_helper(A, 0, len(A) -1)

def func8_helper(A, i, j):
    #Input: An array A and two indices i and j.
    n = j-i+1
    if n ==2:
        if A[i] > A[j]:
            # Swap A[i] and A[j]
            A[i], A[j] = A[j], A[i]
        elif n > 2:
            m = n // 3
            func8_helper(A, i, j - m)
            func8_helper(A, i + m, j)
            func8_helper(A, i, j - m)
        return A
```

```python
#Sample function call: func8([3,1,2,7,8,6])
A = [3,1,2,7,8,6]
print(func8(A))

import timeit
sec = timeit.timeit(lambda :func8(A.copy()), number=1)
print(f"Execution time: {sec*1000:.6f} ms")

#the function performs the recursice sorting of an array
```

```python
# Exercise 7
def func12(arr, k):
    arr_len = len(arr)
    k = k % arr_len
    # Initialize a new array of the same length as 'arr'
    custom_arr = [0] * arr_len

    for i in range(arr_len):
        mystery_index = (i + k) % arr_len
        custom_arr[mystery_index] = arr[i]
    return custom_arr

```

```python
#Sample function call: func12([1, 2, 3, 4, 5], 2)
print(func12([1, 2, 3, 4, 5], 2))
import timeit
sec = timeit.timeit("func12(arr,k)", globals=globals(), number=1)
print(f"Execution time: {sec*1000:.6f} ms")
```

```python
# write a function named as sum_arr to find sum of elements in given array
def sum_arr(arr):
  sum = 0
  for num in arr:
    sum += num
  return sum
```

```python
#call the function
#arr = [12, 3, 4, 15,45,78,450]
#print(sum_arr)
arr = [12, 3, 4, 15,45,78,450]
print(sum_arr(arr))
import timeit
sec = timeit.timeit("sum_arr(arr)", globals=globals(), number=1)
print(f"Execution time: {sec*1000:.6f} ms")
```

```python
#write a function named as max_min to find minimum (or maximum) element in an array
def max_min(arr):
    choice = input("Do you want to find 'max' or 'min'? ").strip().lower()
    if choice == "max":
        result = arr[0]
        for num in arr[1:]:
            if num > result:
                result = num
        print("Maximum element:", result)
        return result
    elif choice == "min":
        result = arr[0]
        for num in arr[1:]:
            if num < result:
                result = num
        print("Minimum element:", result)
        return result
```
